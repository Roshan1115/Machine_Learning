```python
#load famous iriss dataset
from sklearn import datasets
from sklearn.neighbors import KNeighborsClassifier

iris = datasets.load_iris()

features = iris.data
labels = iris.target

# training classifier
clf = KNeighborsClassifier()
clf.fit(features, labels)

random_test_feature_1 = [1,1,1,1]
random_test_feature_2 = [31, 0.6, 0.5, 0.5]
predicted = clf.predict([ random_test_feature_1, random_test_feature_2 ])
print(predicted)
```
