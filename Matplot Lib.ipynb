```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0,1,2,3])
y = np.array([3,8,1,10])

#format to pass line and marker parameter in a single string
# "marker | line | color". here marker = o, line = --, color = r

# ms = marker size
# mec = marker edge color
# mfc = marker face color

plt.plot(x,y, 'o--r', ms=10, mec ='blue', mfc = 'green')

plt.xlabel("It is x axis")
plt.ylabel("It is y axis")
plt.title("My first Plot Graph")

# default axis value is both x and y
plt.grid(color='green', linestyle='--', linewidth='1', axis='x')

```

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y, 'o-.')

plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

# s = size, c = color

plt.scatter(x, y, c='hotpink', s=150)

x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y, c = 'green')

plt.show()
```

```python
x = np.array([2,5,8])
y = np.array([100,105,84])

colors = np.array(["red","green","blue"])
sizes = np.array([100, 200, 300])

plt.scatter(x,y, c=colors, s=sizes, alpha=0.5)
```

```python
# bars


x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

#virtial
plt.bar(x, y, color='green', width=0.5)
plt.show()

#horizental
plt.barh(x,y, color='hotpink', height=0.5)

# width in virtical == height in horizental
```

```python
# histogram

# in normal function, 150 is mean, 5 is standard deviation, 200 total elements

x = np.random.normal(150, 5, 200)

plt.hist(x)
plt.show() 
```
