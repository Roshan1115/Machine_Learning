```python
import numpy as np

p = np.array([ [1,7], [2,1], [3,2] ])
print("p :",p)

q = np.array([ [1,1], [1,-1], [1,0] ])
print("q :",q)

# addition
c = np.add(p,q);
print("addition :", c)

r = np.array([ [1,3,1], [1,0,1] ])
print("r :", r)

# multiplication
d = np.dot(p, r)
print("multiplication of p and r :", d)

# determinent
det = np.linalg.det(d)
print("determinent of d is", det)

# inverse
m = np.array([ [1,-1], [2,3] ])
print("inverse of m : ", np.linalg.inv(m))
```
