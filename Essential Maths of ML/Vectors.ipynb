```python
import numpy as np

v = np.array([1, -1, 2])
w = np.array([2, 5, 2])

# Addition subtraction
print("Addition :", v + w)
print("Addition :", v - w)

# scalar multiplication
print("scalar multiplication : " , v * 3)

# length
print("length of v :", np.linalg.norm(v))

# dot product
s = np.dot(v,w);
print("dot product of v and w :", s)
```
