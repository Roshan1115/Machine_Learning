#Heart Desease Dataset Experiment


```python
hd = pd.read_csv('/content/drive/MyDrive/Machine_Learning/Udemy_ML_Course/Matplotlib/heart-disease.csv')
```

```python
hd.head()
```

```python
hd['age'].plot.hist(bins=10);
```

```python
hd.plot.hist()
```

```python
over_50 = hd[hd['age'] > 50]
over_50
```

```python
#pyplot methode
over_50.plot(x = 'age', y = 'chol', c = 'target', kind='scatter')
```

```python
# Object Oriented methode mixed with pyplot methode
fig, axis = plt.subplots(figsize = (10, 6))
over_50.plot(kind='scatter',
             x = 'age',
             y = 'chol',
             c = 'target',
             ax = axis);
```

```python
# Object Oriented mehtode
fig, ax = plt.subplots(figsize=(10,6))

# plot
scatter = ax.scatter(x = over_50['age'],
                     y = over_50['chol'],
                     c = over_50['target'])

# Customize the axis
ax.set(title="Cholesterol Level over ages",
       xlabel='Age',
       ylabel='Cholesterol Level');


# Add legend
ax.legend(*scatter.legend_elements(), title='Heaert Desease');

# add horizental line
ax.axhline(over_50['chol'].mean(),
           linestyle='--');
```

```python
over_50.head()
```

```python
# subplot of chol, age and thalach
fig, (ax0,ax1) = plt.subplots(nrows = 2,
                              ncols= 1,
                              figsize = (10, 10),
                              sharex = True);

# add data to first axis
scatter = ax0.scatter(x = over_50['age'],
                      y = over_50['chol'],
                      c = over_50['target']);

ax0.set(title="Cholesterol over age",
        ylabel = 'Chosterol')
ax0.legend(*scatter.legend_elements(), title='Heaert Desease')
ax0.axhline(over_50['chol'].mean(), linestyle="--");


# add data to second axis
scatter = ax1.scatter(x = over_50['age'],
                      y = over_50['thalach'],
                      c = over_50['target']);

ax1.set(title="Max Heart Rate over age",
        xlabel='Age',
        ylabel = 'Max Heart Rate'); 
ax1.legend(*scatter.legend_elements(), title='Heaert Desease')
ax1.axhline(over_50['thalach'].mean(), linestyle="--");

# set property of whole figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight='bold');
```
