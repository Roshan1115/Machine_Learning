#Nut Butter Example

```python
np.random.seed(0)
sales = np.random.randint(20, size = (5, 3))
sales
```

```python
weekly_sales = pd.DataFrame(sales, index = ['Mon', 'Tues', 'Wed', 'Thurs', 'Fri'],
                            columns=['Almond Butter', 'Peanut Butter', 'Cashew Butter'])
weekly_sales
```

```python
price = np.array([10, 8, 12])
price
```

```python
price.shape
```

```python
price.reshape(1,3)
```

```python
butter_price = pd.DataFrame(price.reshape(1,3), index = ['Price'], 
                            columns=['Almond Butter', 'Peanut Butter', 'Cashew Butter'])
butter_price
```

```python
weekly_sales
```

```python
butter_price.shape
```

```python
weekly_sales.T
```

```python
total_sales = butter_price.dot(weekly_sales.T)
total_sales
```

```python
total_sales.shape
```

```python
weekly_sales['Total Sales'] = total_sales.T
weekly_sales
```

#Sorting Arrays

```python
random_array5 = np.random.randint(10, size = (3,5))
random_array5
```

```python
np.sort(random_array5)
```

```python
np.argsort(random_array5) # return the original indices of element after sorted
```

```python
random_array5
```

```python
np.argmin(random_array5) # return index of smallest element from entire array
```

```python
np.argmax(random_array5)  # return index of largest element from entire array
```

```python
np.argmin(random_array5, axis=0)  # return index of smallest element in each column.
```

```python
np.argmax(random_array5, axis=1)   # return index of largest element in each row.
```
